#### Correlation Between the Columns
Can we find any significant relationships between the columns.

```python
corr = data.corr()
corr
```

```python
#Let's see the correlations
fig = plt.figure(figsize = (8,8))
plt.matshow(corr, cmap = 'RdBu', fignum = fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation = 'vertical');
plt.yticks(range(len(corr.columns)), corr.columns);
```

Blue means high correlation. Negative correlation is dark red. For instance "Profit" has a high correlation between "Unit_Cost" and "Unit_Price".

#### Column Wrangling
We will create new columns or modify existing ones.

```python
#I will add a new "Revenue_per_Age" column
data['Revenue_per_Age'] = data['Revenue'] / data['Customer_Age']
data['Revenue_per_Age'].head()
```

```python
data['Revenue_per_Age'].plot(kind='density', figsize = (14,6))
```

```python
data['Revenue_per_Age'].plot(kind='hist', figsize = (14,6))
```

#### I will add and calculate a new "Calculated_Cost" column
Formula : Order Quantity * Unit Cost

```python
data['Calculated_Cost'] = data['Order_Quantity'] * data['Unit_Cost']
data['Calculated_Cost'].head()
```

```python
#How many rows had a different value
(data['Calculated_Cost'] != data['Cost']).sum()
```

```python
#Let's see the relationship between 'Cost' and 'Profit' using scatterplot
data.plot(kind = 'scatter', x = "Cost", y = "Profit", figsize = (6,6))
```

#### I will add and calculate a new "Calculated_Revenue" column
Formula: Cost + Profit

```python
data['Calculated_Revenue'] = data['Cost'] + data['Profit']
data['Calculated_Revenue'].head()
```

```python
(data['Calculated_Revenue'] != data['Revenue']).sum()
```

```python
data['Revenue'].plot(kind = 'hist', bins = 100, figsize = (14,6))
```
